# Financing - Decision Trees and Random Forests Project

This project is an exploration of publicly available data from LendingClub.com. Lending Club connects people who need money (borrowers) with people who have money (investors). As an investor you would want to invest in people who showed a profile of having a high probability of paying you back. This model will help predict whether or not an individual will payback a loan based off of profile data collected on individuals throughout the life of various loans. 

This data is from 2007-2010, before Lending Club went public. The dataset contains the following information:
* credit.policy: 1 if the customer meets the credit underwriting criteria of LendingClub.com, and 0 otherwise.
* purpose: The purpose of the loan (takes values "credit_card", "debt_consolidation", "educational", "major_purchase", "small_business", and "all_other").
* int.rate: The interest rate of the loan, as a proportion (a rate of 11% would be stored as 0.11). Borrowers judged by LendingClub.com to be more risky are assigned higher interest rates.
* installment: The monthly installments owed by the borrower if the loan is funded.
* log.annual.inc: The natural log of the self-reported annual income of the borrower.
* dti: The debt-to-income ratio of the borrower (amount of debt divided by annual income).
* fico: The FICO credit score of the borrower.
* days.with.cr.line: The number of days the borrower has had a credit line.
* revol.bal: The borrower's revolving balance (amount unpaid at the end of the credit card billing cycle).
* revol.util: The borrower's revolving line utilization rate (the amount of the credit line used relative to total credit available).
* inq.last.6mths: The borrower's number of inquiries by creditors in the last 6 months.
* delinq.2yrs: The number of times the borrower had been 30+ days past due on a payment in the past 2 years.
* pub.rec: The borrower's number of derogatory public records (bankruptcy filings, tax liens, or judgments).

____
**Categorical Features**

The purpose column is categorical (type object). It will need to be transformed using dummy variables so sklearn will be able to understand them. This can be done in one step using pd.get_dummies.

```python
final_data = pd.get_dummies(loans, columns=['purpose'], drop_first=True)
```

```python
final_data.info()
```

___
**Training Decision Tree Model**

This is a supervised learning model, so the data will be split into training and testing sets. The model chosen is a Decision Tree, which is a simple algorithm that uses recursive binary splitting to segment the feature space into a number of simple regions. For this model the prediction is not fully paid and all the remaining columns comprise the feature space.

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid', axis=1)
y = final_data['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

___
**Predictions and Evaluation**

What follows is the predictions from the test data. Again this is supervised learning so there is already true output variables that are stored in the y_test dataset that the predicted values can be compared against to determine how well the model performed.

Two reports are created to determine the effectiveness of the model. First, the confusion matrix, graphically a confusion matrix can be interpreted as follows:

|   n=3161   | Pred No | Pred Yes |
|------------|---------|----------|
|  Actual No |  2167   |   499    |
| Actual Yes |  381    |   114    |

The model mislabeled 880 data points, using this information with the classification report it is clear how the percentages are arrived at. For example the model predicted 2,548 observations to be 0 (No), out of those only 2167 were actually 0 for a precision of 2167/2584 = 0.85. Looking at recall, the percentage is generated off of all the test points that are actually 0, according to the classification report that is 2666 (found under the support column) 2167/2666 = 0.81.

In general the model performed poorly and can be improved, by using a random forest algorithm to decrease the variance in the model.

```python
predict = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predict))
print('\n')
print(classification_report(y_test, predict))
```

___
**Random Forest Model**

This will be trained with the same dataset used for the decision tree to make an effective comparison between the two models. The basic difference between random forests and decision trees is that random forests use a technique called bagging with a random sample of predictors to generate multiple decision trees and then combine these trees to form one, using the average of the predictions.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train, y_train)
```

```python
predict_rfc = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, predict_rfc))
print('\n')
print(classification_report(y_test, predict_rfc))
```

___
**Conclusion**

Based on the classification report and confusion matrix it is clear that the random forest performed better than the decision tree. This was expected. There was a reduction of mislabeled data points by 385. However looking at the f1-score which is a weighted average of the precision and recall the change wasn't that drastic. In addition, the recall for class 1 dropped from .23 for the decision tree to .02 for the random forest model. So, depending on what parameter is being optimized, will determine the choice in the algorithm.
