**Training K Means Cluster Model**

This is an unsupervised learning model, so the data will only have features and conclusions will be drawn from them. The model chosen is a K Means Cluster model, which is an algorithm that seeks to partition the observations into a prespecified number of clusters K (in this case 2). It does this by randomy assigning a number from 1 to K to each observation, these are initial cluster assignments. For each of the clusters it computes the centroid and then reassigns each observation based on the cluster that is closest until the observations stop changing.

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(data.drop('Private', axis=1))
```

The model converged in 17 iterations

```python
kmeans.cluster_centers_
```

**Model Evaluation**

In general this is not a step when using unsupervised learning algorithms, the main reason for using unsupervised learning is to generate analysis of the data. However here the project is looking at the effectiveness of K means clustering for separating out the students into private and public classes. 

The first method of evaluation was recreating the two histograms above using the labels generated from the model. It looks like the model overlabelled students as private (private = 1), this is quite evident in the histogram that shows frequency of out of state tuition prices. The lower price ranges see a higher frequency of private students from the model.

To confirm this a confusion matrix and classification report are generated, comparing the original classification of private and the model classifications. The model mislabelled 172 data points for a weighted f1-score of 75, this is poor classification.

```python
labels = pd.DataFrame(kmeans.labels_, index=data.index)
cluster_df = pd.concat([data, labels], axis=1)
```

```python
g = sns.FacetGrid(cluster_df, hue=0, height=6, aspect=2, palette='coolwarm')
g.map(plt.hist, 'Outstate', alpha=0.7, bins=30)
plt.ylabel('Number of Students')
plt.legend()
plt.title('Frequency of Out of State Tuition Prices for K Means Clusters 0 and 1')
```

```python
g = sns.FacetGrid(cluster_df, hue=0, height=6, aspect=2, palette='coolwarm')
g.map(plt.hist, 'Grad.Rate', alpha=0.7, bins=30)
plt.ylabel('Number of Students')
plt.legend()
plt.title('Frequency of Graduation Rates for K Means Clusters 0 and 1')
```

```python
#Function to convert private column to 0 and 1 instead of No and Yes.
def converter(label):
    if label == 'No':
        return 0
    else:
        return 1
```

```python
data['Private_Convert'] = data['Private'].apply(converter)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(data.Private_Convert, cluster_df[0]))
print('\n')
print(classification_report(data.Private_Convert, cluster_df[0]))
```
