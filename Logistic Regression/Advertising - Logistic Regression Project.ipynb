# Advertising Project - Logistic Regression

In this project I'm working with a fake advertising data set, to determine whether or not a particular internet user clicked on an Advertisement. I create a model using logistic regression that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': Customer time on site in minutes
* 'Age': Cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

___
**Training and Testing**

This is a supervised learning model, so the data will be split into training and testing sets. The model chosen is Logistic Regression, which is a simple algorithm for predicting a qualitative response from various predictor variables. For this model the prediction is binary, Clicked on Ad and, the predictors are Daily Time Spent on Site, Age, Area Income, Daily Internet Usage, Male.

```python
from sklearn.model_selection import train_test_split
```

```python
X = ads[['Daily Time Spent on Site', 'Age','Area Income', 'Daily Internet Usage','Male']]
y = ads['Clicked on Ad']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression(solver='liblinear')
logmodel.fit(X_train, y_train)
```

___
**Predictions and Evaluation**

First the prediction set is made from the test data separated out from the dataset. Then the model is evaluated using a classification report. This model ended up performing very well. Precision and recall are the most widely used metrics for how well a system performs.

Interpreting the results below the system has 87% and 96% precisions for not clicked on ad (0) and clicked on ad (1), with an average of 91%. Using not clicked on ad (0) as an example what precision says is that out of all the points the system labeled 0, 87% of them were actually 0. Looking at recall the system has 96% and 86% recall for 0 and 1 respectively, with an average of 91%. Again, using 0 as an example what recall says is that out of all the data points that are actually 0 the system labeled 96% of the data points correctly.


```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```
